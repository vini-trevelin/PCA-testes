Teste do impacto de parametros no PCA

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from matplotlib.pyplot import imread
```

```python
img = imread("frame12.jpg")
# garantir que ta tudo em int
img = img.astype(np.uint8)
# normalizando
img = img/255
img = img.mean(axis=2)
# coloco em cinza (ainda n descobri pq mas muita gente faz isso)
plt.imshow(img,cmap="gray")
```

```python
# função que comprime a imagem em % (numero de components)
# define a variancia
def transforma (porcentagem):
    porcentagem = porcentagem/100
    tswizzle_pca = PCA(n_components=porcentagem).fit(img)
    transformado = tswizzle_pca.transform(img) 
    projecao = tswizzle_pca.inverse_transform(transformado)
    plt.imshow(projecao, cmap="gray")
```

```python
transforma(50)
```

Código teste-> pega imagem e transforma em 3 np.arrays rgb

```python
import PIL
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
image_array = PIL.Image.open("teste_RGB.png")
arr = np.array(image_array)

r,g,b = cv.split(arr)
l, c = r.shape #pego numero de linhas e colunas da img
l,c
# cv.imshow("red",r)
# cv.imshow("green",g)
# cv.imshow("blue",b)
 
```

Calculando media dos valores RGB em toda img (soma de rgb dos pixels / numero de pixels)(para talvez usar como parametro)

```python
red_media = np.sum(r)/(l*c)
green_media = np.sum(g)/(l*c)
blue_media = np.sum(b)/(l*c)
red_media,green_media,blue_media

```

Valor de RED menor que GREEN/BLUE pois a img tem mais colunas que linhas, o que aumenta o número de pixels em G e B e não em R
